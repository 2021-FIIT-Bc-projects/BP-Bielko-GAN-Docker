```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fastRandomPointFromDistance(origin, distance, n_dims):
    distance_squared = distance ** 2
    a = np.random.normal(size=n_dims)
    a /= sum(a)
    a *= distance_squared
    a = np.sqrt(a)
    for i, coord in enumerate(a):
        a[i] *= np.random.choice([1, -1])
    return a
```

```python
fastRandomPointFromDistance(np.asarray([0,0]), 1, 100)
```

```python
points = [fastRandomPointFromDistance(np.asarray([0,0]), 3, 3) for i in range(1000)]
pointsx = [point[0] for point in points]
pointsy = [point[1] for point in points]
pointsz = [point[2] for point in points]

fig = plt.figure(figsize=(5, 5))
ax = plt.axes(projection="3d")

ax.scatter(pointsx, pointsy, pointsz)
plt.show()
```
